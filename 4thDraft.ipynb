### **Data Cleaning**

Dropping Duplicate rows

```python
df.drop_duplicates(inplace=True)
```

Handling Null Values

```python
df['reviews_per_month'].fillna(0,inplace=True)
df['last_review'].fillna(0,inplace=True)

columns_to_fill = ['name', 'host_name']
df[columns_to_fill]=df[columns_to_fill].fillna('Unknown')
```

Naming columns appropriately

### **Popular neighbourhod based on reviews**

Top Neighbourhood_group based on reviews

```python
reviews_grp=df.groupby('neighbourhood_group')['number_of_reviews'].sum().reset_index()

reviews_grp
```

```python
fig,ax=plt.subplots()
sns.barplot(data=reviews_grp,x='neighbourhood_group',y='number_of_reviews',color='#63BAB6',width=0.6)
ax.set_xlabel('Neighborhood Group')
ax.set_ylabel("Total Number of Reviews")
ax.set_title('Number of Reviews by Neighborhood Group')
fig.text(1,0.04,'X',ha='center')
fig.text(0.06,1,'Y',va='center',rotation='vertical')

#ax.set_xticks
plt.show()
```

Famous neighbourhood group is Broklyn

```python
brooklyn_df=df.groupby('neighbourhood_group').get_group('Brooklyn')
brooklyn_df.head()
```

```python
neighbourhoods_of_brooklyn=brooklyn_df.groupby('neighbourhood')['number_of_reviews'].sum().reset_index()
top_neighbourhoods_of_brooklyn=neighbourhoods_of_brooklyn.sort_values('number_of_reviews',ascending=False).head(10)
top_neighbourhoods_of_brooklyn
```

```python
fig,ax=plt.subplots()
sns.barplot(data=top_neighbourhoods_of_brooklyn,x='neighbourhood',y='number_of_reviews',color='#63BAB6',width=0.6)
ax.set_xlabel('Neighborhood ')
ax.set_ylabel("Total Number of Reviews")
ax.set_title('Number of Reviews by Neighborhood Group')
fig.text(1,0.04,'X',ha='center')
fig.text(0.06,1,'Y',va='center',rotation='vertical')
plt.xticks(rotation=90)

#ax.set_xticks
plt.show()
```

Most popular Neighbourhood group is Broklyn based on reviews
And most popular neighbour from Brooklyn is Bedford-Stuyvesant

### **Top hosts per each room type**

```python
roomtype_1=df.groupby('room_type').get_group('Private room')
roomtype_2=df.groupby('room_type').get_group('Entire home/apt')
roomtype_3=df.groupby('room_type').get_group('Shared room')
```

```python
hosts_roomtype_1=roomtype_1.groupby('host_name')['number_of_reviews'].sum().reset_index()
top_hosts_roomtype_1=hosts_roomtype_1.sort_values('number_of_reviews',ascending=False).head()
top_hosts_roomtype_1
```

```python
hosts_roomtype_2=roomtype_2.groupby('host_name')['number_of_reviews'].sum().reset_index()
top_hosts_roomtype_2=hosts_roomtype_2.sort_values('number_of_reviews',ascending=False).head()
top_hosts_roomtype_2
```

```python
hosts_roomtype_3=roomtype_3.groupby('host_name')['number_of_reviews'].sum().reset_index()
top_hosts_roomtype_3=hosts_roomtype_3.sort_values('number_of_reviews',ascending=False).head()
top_hosts_roomtype_3
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# axes[0].bar(top_hosts_r1['host_name'],top_hosts_r1['number_of_reviews'])
sns.barplot(data=top_hosts_roomtype_1,x='host_name',y='number_of_reviews',ax=axes[0],palette='flare',width=0.6)
axes[0].set_title('Top 5 Hosts of Room type-Private room')

sns.barplot(data=top_hosts_roomtype_1,x='host_name',y='number_of_reviews',ax=axes[1],palette='flare',width=0.6)
axes[1].set_title('Top 5 Hosts of Room type-Entire home/apt')


sns.barplot(data=top_hosts_roomtype_1,x='host_name',y='number_of_reviews',ax=axes[2],palette='flare',width=0.6)
axes[2].set_title('Top 5 Hosts of Room type-Shared room')

plt.tight_layout()
plt.show()
```
