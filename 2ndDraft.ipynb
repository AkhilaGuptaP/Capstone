```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets_2/Airbnb NYC 2019.csv')
df.head().T
```

```python
df.isna().sum()
```

```python
df['calculated_host_listings_count']
```

```python
df['reviews_per_month'].fillna(0,inplace=True)
df.info()
```

```python
columns_to_fill = ['last_review','name', 'host_name']

df[columns_to_fill]=df[columns_to_fill].fillna('NA')

```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python

df.head().T
```

```python
#Most popular one as per reviews
plt.figure(figsize=(20,7))
sns.barplot(data=df.head(100),x='name',y='number_of_reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
reviews_grp=df.groupby('neighbourhood_group')['number_of_reviews'].sum()

reviews_grp

```

```python
reviews_grp=df.groupby('neighbourhood_group')['number_of_reviews'].sum().reset_index()

reviews_grp
```

```python
fig,ax=plt.subplots()
ax.bar(reviews_grp['neighbourhood_group'],reviews_grp['number_of_reviews'],color='green')

ax.set_xlabel('Neighborhood group')
ax.set_ylabel('Total Number of reviews')
ax.set_title('Number of Reviews by Neighborhood Group')
fig.text(0.5,0.04,'X',ha='center')
fig.text(0.04,0.5,'Y',va='center',rotation='vertical')
plt.show()
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='number_of_reviews',palette='flare')
plt.xlabel('Neighborhood Group')
plt.ylabel('Total Number of Reviews')
plt.title('Number of Reviews by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```
